```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

def plot_gantt_chart(routes, routes_legends):
    fig, gnt = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed
    
    # Set Y-axis limits based on the number of routes
    gnt.set_ylim(0, 50 * len(routes))
    
    # Define colors for different routes
    route_colors = ['tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

    # Convert time strings to datetime objects
    start_times = [datetime.strptime(leg[2], '%H:%M') for route in routes for leg in route]
    end_times = [datetime.strptime(leg[3], '%H:%M') for route in routes for leg in route]

    # Define time range for X-axis
    min_time = min(start_times)
    max_time = max(end_times)
    gnt.set_xlim(min_time - timedelta(hours=1), max_time + timedelta(hours=1))

    # Set labels for X-axis and Y-axis
    gnt.set_xlabel('Time')
    gnt.set_ylabel('String')
    
    # Setting Y-axis ticks and labels for each route
    y_ticks = []
    y_labels = []
    for i, route in enumerate(routes):
        y_ticks.append(25 + i * 50)
        y_labels.append(f'String {i+1}')
    gnt.set_yticks(y_ticks)
    gnt.set_yticklabels(y_labels)

    # Setting graph attributes
    gnt.grid(True)
    
    # Plotting each leg of the routes
    for i, route in enumerate(routes):
        for j, leg in enumerate(route):
            start_time = datetime.strptime(leg[2], '%H:%M')
            end_time = datetime.strptime(leg[3], '%H:%M')
            # Convert datetime objects to numerical values
            start_num = mdates.date2num(start_time)
            end_num = mdates.date2num(end_time)
            # Plotting the bars
            gnt.broken_barh([(start_num, end_num - start_num)], (25 + i * 50 - 10, 20), facecolors=route_colors[i], edgecolor='black')

    # Adding legends for route colors
    legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in route_colors[:len(routes)]]
    legend_labels = [f'String {i+1} : {routes_legends[i]}' for i in range(len(routes))]
    gnt.legend(legend_handles, legend_labels, loc='lower right', bbox_to_anchor=(1, 1))

    # Formatting x-axis ticks as time
    gnt.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()

# Example data
routes = [[('BOM', 'DEL', '05:00', '07:15', 120), ('DEL', 'AMD', '08:45', '10:30', 260), ('AMD', 'BOM', '12:45', '14:30', 95)]
          , [('BOM', 'DEL', '06:00', '08:15', 35), ('DEL', 'AMD', '09:10', '10:55', 100), ('AMD', 'BOM', '11:45', '13:30', 75)]]

routes_legends = [[('BOM', 'DEL', '05:00', '07:15'), ('DEL', 'AMD', '08:45', '10:30'), ('AMD', 'BOM', '12:45', '14:30')]
                  , [('BOM', 'DEL', '06:00', '08:15'), ('DEL', 'AMD', '09:10', '10:55'), ('AMD', 'BOM', '11:45', '13:30')]]

plot_gantt_chart(routes, routes_legends)

plt.savefig("Test Case 1 - gantt.png")
```
